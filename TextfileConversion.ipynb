Compare the performance (time taken) of C, C++, Java, R, and Python programs for
1. Convert 200MB, 400 MB, 600 MB, 800 MB, and 1000MB text files to upper case.

```python
import pandas as pd
import matplotlib.pyplot as plt
d={'Python':[5.02,7.84,11.84,15.82,20.57],'C':[7.79,16.03,23.91,32.57,40.36]}
table=pd.DataFrame(d,index=['200MB','400MB','600MB','800MB','1000MB'])
print(table)

a1=[5.02,7.84,11.84,15.82,20.57]
a2=[7.79,16.03,23.91,32.57,40.36]
x=[200,400,600,800,1000]
plt.plot(x,a1,label='Python',color='blue')
plt.plot(x,a2,label='C',color='red')
plt.xlabel("File_size(MB)")
plt.ylabel("Time_taken(Sec)")
plt.title("Performance Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
